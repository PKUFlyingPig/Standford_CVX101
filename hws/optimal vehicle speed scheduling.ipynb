```python
z = cp.Variable((n,1))
t = cp.multiply(z,d)
t_cum = cp.cumsum(t)
f = a*cp.inv_pos(z) + c*z + b
```

```python
obj = cp.Minimize(f.T@d)
constraints = [z >=0, 1/smax <= z, z <= 1/smin, tau_min <= t_cum, t_cum <= tau_max]
prob = cp.Problem(obj, constraints)
prob.solve()
print("status :", prob.status)
print("optimal value :", prob.value)
```
